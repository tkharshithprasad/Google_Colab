```
#3. Add the n number of names in a list and print them alphabetically and reverse alphabetically.
b=[]
n=int(input("ENTER range"))
for i in range(n):
    b.append(input(""))
b.sort()
print(b)
b.reverse()
print(b)
```

```


#4. Print  perfect squares and divisible by 5 between 500 and 1000( (both inclusive) using list comprehension
import math
mul=[]
div=[]
mul=[i for i in range(500,1001) if i%5==0]
for i in mul:
    m=math.sqrt(i)
    if (m-math.floor(m)==0):
        div.append(i)
print(div)

```
