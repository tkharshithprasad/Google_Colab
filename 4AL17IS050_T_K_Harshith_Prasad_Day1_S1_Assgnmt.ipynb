```
def cylinder(r,h):
    a=2*3.14*r*h
    print("Cylinder area is {0}".format(a))
    return
def cone(r,h):
    a=1/3*3.14*r*r*h
    print("Cone area is %0.3f"%(a))
    return

b,c=4,6
cylinder(b,c)
cone(b,c)
```

```
def eo(r):
  if(r%2==0):
    print("Even")
  else:
    print("Odd")
def fact(n):
  fact=1
  while(n>0):
    fact=fact*n
    n=n-1
  print("Factorial of the number is: ")
  print(fact)
def odd(n):
  i=1
  while(i<=n):
    if(i%2!=0):
      print(i)
    i=i+1

def prime(n):
  for val in range(2, n + 1): 
    if val > 1: 
        for n in range(2, val//2 + 2): 
            if (val % n) == 0: 
                break
            else: 
                if n == val//2 + 1: 
                    print(val) 
  
print("1.Even or Odd")
print("2.Factorial")
print("3.Odd upto N")
print("4.Prime upto N")
ch=int(input("Enter your choice"))
if ch==1:
    r=int(input("Enter no"))
    eo(r)
elif ch==2:
    c=int(input("Enter n"))
    fact(c)
elif ch==3:
    r=int(input("Entee n"))
    odd(r)
elif ch==4:
    c=int(input("Enter n"))
    prime(c)
else:
    print("Invalid")
```
