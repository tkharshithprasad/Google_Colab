```
import operator
n=int(input("Enter no of records"))
d={}
for i in range(1,n+1):
    name= input("Enter name %d"%(i))
    salary=int(input("Enter sal %d"%(i)))
    d[name]=salary
c={} 
c=sorted(d.items(),key=operator.itemgetter(1))
print("sum is ",sum(d.values()))
print("min is",min(d.values()))
print("max is",max(d.values()))
print("average is",sum(d.values())/n)

print(c)
```

```
#1) Python Program to count the total number of charaters (except blank space) in a text file.
from google.colab import drive
drive.mount('/content/drive')
#fname = input("Enter file name: ")
fname = 'drive/My Drive/doc.txt'
count=0
k=0
c,l=0,0
with open(fname, 'r') as f:
    for line in f:
        l=l+1
        for i in line:
            count=count+1
            if(i==" "):
                k=k+1
    c=count-k-(l-1)
    print("Number of characters:")
    print(c)
```
